```python
import numpy as np
import pandas as pd
```

```python
census_raw = pd.read_csv('data/census_raw.csv')
census_raw = census_raw.rename(columns={'Label (Grouping)': 'State'})
census_raw = census_raw.set_index('State')
census_raw.head()
```

```python
transposed = census_raw.transpose()
transposed.head()
```

```python
df = transposed[['Total','Median income (dollars)', 'Mean income (dollars)']] #dropped all columns we won't use
df.head()
```

```python
df = df[~df.index.str.contains('Margin of Error')] #removing margin of error category
df.index = df.index.str.replace('!!Estimate','') #this specifier no longer necessary
df = df.reset_index(names='State')
df.index.name = ''
df.head()
```

```python
df[['State','Category']] = df['State'].str.split('!!', n=2, expand=True) #splitting state into state and category columns
df.columns.name = None
df.head()
```

```python
col = df.pop('Category')
df.insert(1, col.name,col) #reordering column
```

```python
df.to_csv('data/census_cleaned.csv', index=False)
```
